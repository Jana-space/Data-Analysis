```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# URL for the CSV data
url = "https://data.transportation.gov/api/views/6aiz-ybqx/rows.csv?accessType=DOWNLOAD"

# Fetch the data from the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read the CSV data into a DataFrame directly from the response content
    from io import StringIO
    csv_data = StringIO(response.text)  # Convert response text to a file-like object
    df = pd.read_csv(csv_data)
    display(df)
else:
    print("Failed to retrieve data:", response.status_code)
```

### Data Cleaning

```python
# Replace '-' with 'Unknown' in 'user_other' and 'exp_type' columns
df['user_other'] = df['user_other'].replace('-', 'Unknown')
df['exp_type'] = df['exp_type'].replace('-', 'Unknown') 
df['own_supporting'] = df['own_supporting'].replace('-', 'Others')
# Verify if '-' has been replaced
print(df[['user_other', 'exp_type', 'own_supporting']].sample(10))
```

```python
# Drop rows with missing values in 'chained_value' 
df = df.dropna(subset=['chained_value'])
# Calculate deflator for rows where it is missing
df.loc[df['deflator'].isnull(), 'deflator'] = (
    (df['chained_value'] / df['value']) * 100
)
#checking for outliers, correctness, accuracy, validity, and incositency
df[['year','value', 'deflator', 'chained_value']].describe()
#evenues cannot have negative
```

```python
df[df.value <0]
"""In financial data, particularly for income statements, net interest income (expense)
 is a measure that reflects the difference between interest income and interest expenses.
   When we encounter a negative value in net interest income (expense), it typically 
   indicates that interest expenses exceeded interest income, resulting in a net interest expense
   rather than net interest income.A negative net interest income (expense) would mean that Amtrak
     paid more in interest on its debt than it earned from interest on its assets. 
     This is commonly referred to as net interest expense. therefore to clean the data we will reclassify
       the cashflow as expenditure and convert the valuses into postive"""

# Step 1: Identify and convert negative values to positive in 'value' and 'chained_value'
df.loc[df['value'] < 0, 'value'] *= -1  # Multiply negative values by -1 to make them positive
df.loc[df['chained_value'] < 0, 'chained_value'] *= -1  # Do the same for chained_value

# Step 2: Change cash_flow from 'Revenue' to 'Expenditure' for records that originally had negative values
df.loc[df['value'] < 0, 'cash_flow'] = 'Expenditure'
```

## Data Quality and Pre-processing 

First, preliminary explorations of dataset also pointed to several quality concerns, which needed to be addressed to ensure the soundness of the analysis. It has 2,546 records within 16 columns and factors such as financial ratios and categorical variables. When assessing the initial stage of missing data, 9 values were lacked in the chained_value and 741 lacked in the deflator. Such limitations would compromise the data quality; to overcome these gaps, we developed a cleaning strategy that sustains statistical integrity of the data:  

During data cleaning, we used a complex technique to address missing values and transform categorical variables. In cases where some of these deflator values were missing, we estimated them using formula based on the ratios comparing nominal and chained values used in the study. The standardization of categorical variables involved replacing ambiguous entries (marked with '-') with more meaningful classifications: 'Unknown' for user_other and exp_type columns, and 'Others' for the own_supporting column. It means that although this standardization enhanced the datasetâ€™s analytical value in most cases, it did not distort its essential nature.  

First of all, one can observe an additional feature inherent in our analysis concerned with the management of negative values within the established dataset. Instead of eliminating these entries directly, it was incumbent to appreciate them as concerning net interest expenses. By following the correct flow of accounting, these negative values were converted into positive amounts and shifted from the revenue account to the expenditure account, ensuring the historical facts are not distorted while enhancing the usability of the dataset for analysis.
