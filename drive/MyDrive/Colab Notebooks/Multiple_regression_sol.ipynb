```python
# Reading the data
df=pd.read_csv('FuelConsumption.csv')
```

```python
#separate the other attributes from the predicting attribute
X = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','FUELCONSUMPTION_COMB_MPG']]

#separte the predicting attribute into Y for model training
y=df[['CO2EMISSIONS']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
# importing module
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
# fitting the training data
regr.fit(X_train,y_train)
```

```python
# The coefficients
print('Coefficients:', regr.coef_)
print('Intercept:', regr.intercept_)
```

```python
y_prediction =  regr.predict(X_test)
y_prediction
```

```python
from sklearn.metrics import r2_score
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_prediction-y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_prediction-y_test)**2))
print("R2-score: %.2f" % r2_score(y_prediction,y_test))
```
